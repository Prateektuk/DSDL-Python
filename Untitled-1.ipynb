```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a,b,c=[1,2,3,4,5],[1,4,9,16,25],[1,8,27,64,125]
plt.xlabel('numbers')
plt.title('sqaure & cude')
plt.ylabel('square')
plt.plot(a,b,'c',linewidth=2,marker='d',markersize=5)
plt.plot(a,c,'g',linewidth=4,linestyle='-.')
plt.show()
```

```python
City=['mumbai','delhi','agra','jaipur']
Popul=[1,4,9,16]
plt.bar(City,Popul,width=[.5,0.7,.3,.8],color=['red','b','g','black'])
plt.xlabel('Cities')
plt.ylabel('Population')
plt.show()
```

```python
a,b,c,d=[1.,2.,3.,4.,5.],[1.,4.,9.,16.,25.],[1.5,2.5,3.5,4.5,5.5],[2.,4.,8.,24.,36.]

plt.bar(a,b,label='square')
plt.bar(c,d,label='cube')
plt.legend(loc='upper left')
plt.show()
```
